```python
df.drop('Unnamed: 0', axis = 1 , inplace = True)
```

```python
df.head()
```

```python
sns.boxplot(x = df['wine_servings'], data = df)
```

```python
import numpy as np
```

```python
q3 = np.percentile(df['wine_servings'],75)
q1= np.percentile(df['wine_servings'],25)

IQR = q3-q1

maxi = q3+ 1.5*IQR
mini = q1 - 1.5*IQR

print('IQR : ',IQR,'\n',
      'Q3:   ', q3,'\n',
      'Q1:   ',q1,'\n',
      'MAX:  ', maxi,'\n',
      'MIN:  ', mini)
```

```python
z = np.percentile(df['wine_servings'],75)
```

```python
df['wine_2'] = df['wine_servings'].apply(lambda x: z if x >= 140 else x)
```

```python
sns.boxplot(x = df['wine_2'], data = df)
```

```python
df['name'] = df.continent + ',' + df.country
df.head()
```

```python
df['beer_servings_1'] = df['beer_servings'].apply(lambda x: 0 if x>100 else 1)
```

```python
df.head()
```

```python
df['abc_servings']=df['wine_servings'].apply(lambda x:'high cosumption'if x >100 else 'low consumption')
```

```python
df.head()
```

```python
['high cosumption' if x >100 else 'low consumption' for x in df['wine_servings']]
```

```python
df['country'] = df.country.apply(lambda x: x.upper())
df['a'] = df['country'].apply(lambda x: x.swapcase())
```

```python
df.head()
```

```python
nr = []

for i in df['beer_servings']:
    if i == 0:
        nr.append(df['beer_servings'].mean())
    else :
        nr.append(i)
    
df['beer_servings'] = nr 

```

```python
df['wine_servings'] = df['wine_servings'].replace(0,df['wine_servings'].mean())
```
