```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
plt.style.use('ggplot')
from matplotlib.pyplot import figure
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
df1 = df.dropna()
```

```python
df1.head(5)
```

```python
df1['budget'] = df1['budget'].astype('int64')
df1['gross'] = df1['gross'].astype('int64')
df1['votes'] = df1['votes'].astype('int64')
```

```python
df1.head(5)
```

```python
df['rating'].unique()
```

```python
for col in df1.columns:
    number_missing = np.sum(df1[col].isnull())
    print('{} - {}'.format(col, number_missing))
```

```python
df1.drop_duplicates()
```

```python
df1[['date_release', 'in_state']] = df1['released'].astype(str).str.split("(", expand = True)
df1['date_release'] = df1['date_release'].str[:-1]
df1['in_state'] = df1['in_state'].str[:-1]
df1.head(10)
```

```python
df1[['date_release', 'year_release']] = df1['date_release'].astype(str).str.split(", ", expand = True)

df1.head()
```

```python
#scatter plot of budget vs gross

plt.scatter(x= df1['budget'], y= df1['gross'])
plt.title('Budget vs Gross')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')

plt.show()
```

```python
#regression plot

sns.regplot(x= 'budget', y= 'gross', data = df1, scatter_kws={'color':'purple'}, line_kws={'color':'black'})
```

```python
correlation_metrics = df1.corr(method = 'pearson')
```

```python
correlation_metrics = df1.corr(method = 'pearson')

sns.heatmap(correlation_metrics, annot = True)
plt.show()
```

```python
#Changing to Categorical Data Type

df1_catt = df1

for col_names in df1_catt.columns:
    if(df1_catt[col_names].dtype == 'object'):
        df1_catt[col_names] = df1[col_names].astype('category')
        df1_catt[col_names] = df1[col_names].cat.codes

df1_catt.head()
```

```python
correlation_metrics2 = df1_catt.corr(method='pearson')

sns.heatmap(correlation_metrics2, annot = True)

plt.show()
```

```python
correlation_met = df1_catt.corr()

corr_pairs = correlation_met.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
High_Corr =sorted_pairs[(sorted_pairs)>0.5]
High_Corr
```
